```python
Anime=pd.read_csv('/content/anime_dataset.csv')
```

```python
Anime.info()
```

```python
Anime.shape
```

```python
Anime.isnull().sum()
```

```python
Anime['year'].fillna(Anime['year'].mode()[0],inplace=True)
```

```python
Anime['year'].isnull().sum()
```

```python
Anime['episodes'].fillna(Anime['episodes'].median(),inplace=True)
```

```python
Anime['episodes'].isnull().sum()
```

```python
Anime.head()
```

```python
Anime['title'].value_counts()
```

```python
Anime['genres'] = Anime['genres'].astype(str).str.strip("[]").str.replace("'", "").str.split(", ")
genres_df=Anime.explode('genres')
genre_count=genres_df['genres'].value_counts()
```

```python
Anime['score'].plot(kind='hist',bins=20,edgecolor='black',title='Distribution of Anime Scores',figsize=(5,3))
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

```

```python
Anime['episodes'].plot(kind='hist',bins=20,edgecolor='black',title='Distribution of Anime Episodes',figsize=(5,3))
plt.xlabel('Episodes')
plt.ylabel('Frequency')
plt.show()
```

```python
Anime['year'].plot(kind='hist',bins=20,edgecolor='pink',title='Distribution of Anime Year',figsize=(10,3))
plt.xlabel('Year')
plt.show()
```

```python
top_10=Anime.sort_values(by='score',ascending=False).head(10)
top_10
```

```python
bottom_10=Anime.sort_values(by='score',ascending=True).head(10)
bottom_10
```

```python
Anime_count=Anime.groupby('year').size()

Anime_count.plot(kind='line',color='pink',figsize=(5,5),title='number of Anime per year')
```

```python
Anime_avg=Anime.groupby('year')['score'].mean()
Anime_avg.plot(kind='line',color='pink',marker='o',figsize=(5,5),title='average score per year')
plt.show()
```
